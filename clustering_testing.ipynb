DBSCAN

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, AgglomerativeClustering
import matplotlib.pyplot as plt
from scipy.stats import t, sem
from scipy.cluster.hierarchy import dendrogram, linkage

# First approach: Simple DBSCAN with very relaxed parameters
def try_dbscan(data, features):
    """
    Try DBSCAN with relaxed parameters
    """
    # Prepare data
    X = data[features].values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Try DBSCAN with very relaxed parameters
    dbscan = DBSCAN(eps=1.0, min_samples=2)
    labels = dbscan.fit_predict(X_scaled)
    
    return labels

# Second approach: Hierarchical clustering
def try_hierarchical(data, features, n_clusters=3):
    """
    Try hierarchical clustering
    """
    # Prepare data
    X = data[features].values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Perform hierarchical clustering
    clustering = AgglomerativeClustering(n_clusters=n_clusters)
    labels = clustering.fit_predict(X_scaled)
    
    # Create linkage matrix for dendrogram
    linkage_matrix = linkage(X_scaled, method='ward')
    
    # Plot dendrogram
    plt.figure(figsize=(10, 7))
    dendrogram(linkage_matrix)
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')
    plt.show()
    
    return labels

# Function to analyze clusters (same as before but with error handling)
def analyze_clusters(data, labels):
    cluster_stats = []
    
    for label in sorted(set(labels)):
        mask = labels == label
        cluster_data = data[mask]
        
        stats = {
            'Cluster': f'Cluster {label}',
            'Size': sum(mask),
            'Galaxies': ', '.join(cluster_data['ID']),
            'Mean_t_bur': cluster_data['t_bur'].mean(),
            'Std_t_bur': cluster_data['t_bur'].std(),
            'Mean_MH_bur': cluster_data['MH_bur'].mean(),
            'Std_MH_bur': cluster_data['MH_bur'].std(),
            'Mean_f_M_sfm': cluster_data['f_M_sfm'].mean(),
            'Mean_DoR': cluster_data['DoR'].mean(),
            'Median_DoR': cluster_data['DoR'].median(),
            'Std_DoR': cluster_data['DoR'].std(),
            'Min_DoR': cluster_data['DoR'].min(),
            'Max_DoR': cluster_data['DoR'].max()
        }
        
        if len(cluster_data) > 1:
            dor_data = cluster_data['DoR'].values
            try:
                ci = t.interval(0.95, 
                              len(dor_data)-1,
                              loc=np.mean(dor_data),
                              scale=sem(dor_data))
                stats['DoR_CI_low'] = ci[0]
                stats['DoR_CI_high'] = ci[1]
            except:
                stats['DoR_CI_low'] = np.nan
                stats['DoR_CI_high'] = np.nan
        
        cluster_stats.append(stats)
    
    return pd.DataFrame(cluster_stats)

# Select features for clustering
features = ['t_bur', 't_max', 'MH_bur', 'MH_max', 'f_M_sfm', 't75', 't_fin']

# Try both methods
print("Trying DBSCAN...")
dbscan_labels = try_dbscan(data, features)
print("\nDBSCAN Results:")
print(analyze_clusters(data, dbscan_labels))

print("\nTrying Hierarchical Clustering...")
hierarchical_labels = try_hierarchical(data, features)
print("\nHierarchical Clustering Results:")
print(analyze_clusters(data, hierarchical_labels))

# Visualize results for both methods
def plot_clustering_results(data, labels, method_name):
    plt.figure(figsize=(15, 5))
    
    # Plot 1: t_bur vs MH_bur
    plt.subplot(131)
    scatter = plt.scatter(data['t_bur'], data['MH_bur'], 
                         c=labels, cmap='viridis')
    plt.xlabel('Burst Time (t_bur)')
    plt.ylabel('Metallicity (MH_bur)')
    plt.title(f'{method_name}: Physical Parameters')
    plt.colorbar(scatter)
    
    # Plot 2: DoR vs t_bur
    plt.subplot(132)
    scatter = plt.scatter(data['t_bur'], data['DoR'],
                         c=labels, cmap='viridis')
    plt.xlabel('Burst Time (t_bur)')
    plt.ylabel('Degree of Relicness (DoR)')
    plt.title(f'{method_name}: DoR vs t_bur')
    
    # Plot 3: DoR distribution
    plt.subplot(133)
    for label in sorted(set(labels)):
        cluster_dor = data[labels == label]['DoR']
        plt.boxplot(cluster_dor, positions=[label])
    plt.xlabel('Cluster')
    plt.ylabel('DoR')
    plt.title(f'{method_name}: DoR Distribution')
    
    plt.tight_layout()
    plt.show()

# Plot results
plot_clustering_results(data, dbscan_labels, 'DBSCAN')
plot_clustering_results(data, hierarchical_labels, 'Hierarchical')

# Save results
results = pd.DataFrame({
    'ID': data['ID'],
    'DBSCAN_Cluster': dbscan_labels,
    'Hierarchical_Cluster': hierarchical_labels,
    'DoR': data['DoR'],
    't_bur': data['t_bur'],
    'MH_bur': data['MH_bur'],
    'f_M_sfm': data['f_M_sfm']
})
results.to_csv('clustering_comparison.csv', index=False)
```

Hierarchial Clustering

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt


# Method 1: Simple binning by DoR
def bin_by_dor(data, n_bins=4):
    """
    Group galaxies into bins based on Degree of Relicness
    """
    # Create bins
    data['dor_bin'] = pd.qcut(data['DoR'], n_bins, labels=[f'Group {i+1}' for i in range(n_bins)])
    
    # Analyze groups
    group_stats = data.groupby('dor_bin').agg({
        'DoR': ['mean', 'min', 'max', 'count'],
        'ID': lambda x: ', '.join(x)
    }).round(3)
    
    return data['dor_bin'], group_stats

# Method 2: Hierarchical grouping based on multiple parameters
def hierarchical_grouping(data, n_groups=4):
    """
    Group galaxies using hierarchical clustering based on key parameters
    """
    # Select features for grouping
    features = ['DoR', 't_bur', 'MH_bur', 'f_M_sfm']
    X = data[features].values
    
    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Perform clustering
    clustering = AgglomerativeClustering(n_clusters=n_groups)
    labels = clustering.fit_predict(X_scaled)
    
    # Add group labels to data
    data['hier_group'] = [f'Group {i+1}' for i in labels]
    
    # Analyze groups
    group_stats = data.groupby('hier_group').agg({
        'DoR': ['mean', 'min', 'max', 'count'],
        'ID': lambda x: ', '.join(x)
    }).round(3)
    
    return data['hier_group'], group_stats

# Execute both methods
n_groups = 4  # Adjust this based on how many spectral groups you want to compare

# Method 1: DoR binning
dor_groups, dor_stats = bin_by_dor(data, n_groups)

# Method 2: Hierarchical grouping
hier_groups, hier_stats = hierarchical_grouping(data, n_groups)

# Plotting
plt.figure(figsize=(15, 5))

# Plot DoR binning results
plt.subplot(121)
for group in sorted(set(dor_groups)):
    mask = dor_groups == group
    plt.scatter(data[mask]['DoR'], data[mask]['t_bur'], label=group)
plt.xlabel('Degree of Relicness (DoR)')
plt.ylabel('Burst Time (t_bur)')
plt.title('Groups by DoR Binning')
plt.legend()

# Plot hierarchical grouping results
plt.subplot(122)
for group in sorted(set(hier_groups)):
    mask = hier_groups == group
    plt.scatter(data[mask]['DoR'], data[mask]['t_bur'], label=group)
plt.xlabel('Degree of Relicness (DoR)')
plt.ylabel('Burst Time (t_bur)')
plt.title('Groups by Hierarchical Grouping')
plt.legend()

plt.tight_layout()
plt.show()

# Print results
print("\nMethod 1: Groups by DoR binning")
print(dor_stats)
print("\nMethod 2: Groups by hierarchical clustering")
print(hier_stats)

# Save groupings to file
results = pd.DataFrame({
    'ID': data['ID'],
    'DoR': data['DoR'],
    'DoR_Group': dor_groups,
    'Hierarchical_Group': hier_groups
})
results.to_csv('galaxy_groups.csv', index=False)
```
